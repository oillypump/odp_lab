```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

# Gunakan localhost karena Master ada di container yang sama
spark = SparkSession.builder \
    .master("spark://spark-nb:7077") \
    .appName("TestApp") \
    .getOrCreate()

print("Spark Session Berhasil Dibuat!")
print("✅ Spark version:", spark.version)
print("✅ Spark master:", spark.sparkContext.master)
print("✅ App ID:", spark.sparkContext.applicationId)
spark.range(10).show()
```

```python
spark.stop()
```
